**Converting Date and Time Formatting**

```python
df["date_added"] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

**Data Visualisation**

```python
sns.countplot(x='type', data=netflix)
fig = plt.gcf()
fig.set_size_inches(10,6)
plt.title('Type')
plt.xlabel('Count')
plt.ylabel('Type')
```

**Rating of the Shows And Movies**

```python
sns.countplot(x='rating', data=netflix)
fig = plt.gcf()
fig.set_size_inches(10, 6)
plt.title('Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
```

**Relation Between Type and Rating**

```python
plt.figure(figsize=(10, 8))
sns.countplot(x='rating', hue='type', data=netflix)
plt.title('Relation Between Type and Rating')
plt.show()
```

**Pie-Chart for the Type: Movie and TV Shows**

```python
lables=['Movie', 'TV Shows']
size=netflix['type'].value_counts()
colors=plt.cm.Wistia(np.linspace(0,1,2))
explode=[0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size, colors=colors, labels=lables, explode=explode, shadow =True, startangle= 90)
plt.title('Distribution of the Type', fontsize = 25)
plt.legend()
plt.show()
```

**Pie-Chart for the Rating**

```python
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%', shadow = True, figsize = (10, 8))
plt.show()
```
